```python
import threading
from bing_image_downloader import downloader
import os
import uuid
import glob
import shutil

def descargar_imagenes(buscar, i):
    # Modifica la consulta de búsqueda para cada hilo
    buscar_modificado = buscar + " " + str(i)
    output_dir = os.path.join('imagenes', buscar_modificado)
    downloader.download(buscar_modificado, limit=3,  output_dir=output_dir, adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
    
    # Renombra y mueve las imágenes descargadas
    for filename in glob.glob(os.path.join(output_dir, '*')):
        new_filename = os.path.join('imagenes', str(uuid.uuid4()) + os.path.splitext(filename)[1])
        shutil.move(filename, new_filename)

    # Elimina la subcarpeta
    os.rmdir(output_dir)

# Crea los hilos
threads = []
for i in range(10):
    thread = threading.Thread(target=descargar_imagenes, args=("avión", i))
    threads.append(thread)

# Inicia los hilos
for thread in threads:
    thread.start()

# Espera a que los hilos terminen
for thread in threads:
    thread.join()
```
